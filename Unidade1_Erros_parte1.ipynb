# Introdução e Erros

Diversos problemas reais são descritos por modelos matemáticos. De forma bem resumida, um modelo matemático é composto por um conjunto de equações que descrevem (ou representam) um problema ou fenômeno físico.

**Modelagem:**

Inicialmente é necessário equacionar o processo que pretendemos estudar, ou seja, decidir quais são as variáveis relevantes e como elas se relacionam, quais são os dados do problema, etc.

Por exemplo, para descrever o movimento de fluidos, em geral são usadas as componentes do campo de velocidade e a pressão, em cada ponto de uma região previamente definida. As leis de conservação (massa, quantidade de movimento, energia, etc) bem como equações constitutivas, são usadas para estabelecer as relações entre as variáveis escolhidas, dentre elas as incógnitas.

Nem sempre conseguimos considerar todos os fenômenos que ocorrem em um desterminado problema. Assim, nesta etapa, algumas hipóteses simplificadoras podem ser consideradas e obtemos o modelo matemático.

**Solução do problema:**

Uma vez estabelecido o conjunto de equações, o próximo passo pe resolver o problema. Nem sempre é possível obtermos soluções por meio de métodos analíticos, onde as soluções são apresentadas por meio de expressões analíticas.

Dessa forma, no contexto da disciplina INE5202, vamos usar métodos numéricos para calcular uma aproximação para a solução desejada.

O processo de solução, por meio de métodos numéricos ou analíticos, pode resultar em uma solução não satisfatória. Neste caso, é necessário avaliar o método de resolução ou até mesmo o modelo matemático.

Em todo este processo, desde a etapa da modelagem são inseridos erros. O conjunto destas incertezas vai afetar a solução final. É necessário termos conhecimento dos tipos de erros envolvidos e como eles se propagam. Algumas fontes de erros são:

*   **Erros Iniciais:**

Inclui incertezas introduzidas na modelagem, medição de parâmetros, condições iniciais, etc. São erros que já existem antes de começarmos o processo de resolução.

A influência destas perturbações iniciais depende da estabilidade do problema. Em alguns casos, a influência deste tipo de erro pode inviabilizar a solução numérica. Estes são problemas conhecidos como mal postos ou mal condicionados, ou seja, são problemas sensíveis a pequenas perturbações nas condições iniciais.

*   **Erros de Truncamento:**

São erros associados ao truncamento (parada) de um processo antes do fim ou quando ele é infinito. Precisamos definir uma aproximação que é obtida após um número finito de passos.

**Exemplo 1.** Calcular $e^x$ com $x=0.5$.

Podemos aproximar $e^x$ calculando

$$e^x = 1 + \sum_{i=1}^{\infty}\dfrac{x^i}{i!}=1 + x + \dfrac{x^2}{2!} + \dfrac{x^3}{3!}+\ldots+\dfrac{x^n}{n!}+\ldots$$

Note que temos uma soma infinita, que precisa ser terminada. Por exemplo,

O que acontece se incluirmos três termos desta soma? Faça a comparação com o valor devolvido para $e^{0.5}$ por uma calculadora científica.


**Nota:** A expressão acima é obtida por meio  da série de Taylor de uma função $f(x)$, é dada por
$$f(x) = f(x_0) + \sum_{n=1}^{\infty}\dfrac{f^{(n)}(x_0)(x - x_0)^n}{n!},$$
sendo $f^{(n)}$ a $n$-ésima derivada de $f$ e $x=x_0+h$, o que significa que $x$ está em uma vizinhança de $x_0$.

**Exemplo 2.** A derivada de uma função pode ser aproximada por,
$$f'(x) \approx \dfrac{f(x+h) - f(x)}{h},$$
com $h$ suficientemente pequeno.

Esta aproximação pode ser observada a partir da definição de derivada e tammbém a partir da Série de Taylor:

$$f(x+h) = f(x) + f^{'}(x)h+\dfrac{f^{''}(x)h^2}{2}+\ldots+\dfrac{f^{(n)}(x)h^n}{n!}+\ldots$$

Note que na expressão acima desconsideramos os termos a partir de $\displaystyle{\dfrac{f^{''}(x)h^2}{2}}$. Como $h$ deve ser suficientemente pequeno (em geral, é menor do que $1$), na notação assintótica o erro cometido nesta aproximação é $O(h^2)$. Isso significa que o erro cometido ao desprezarmos estes termos na Série de Taylor é menor ou igual a $C\cdot h^2$, sendo $C$ uma constante.

Neste caso, queremos aproximar $f'(x)$. A partir da série de Taylor, obtemos

$$f'(x) = \dfrac{f(x+h) - f(x)}{h} + O(h),$$

sendo $O(h)$ o erro cometido na aproximação.


*  **Erros na representação dos números**  

São os erros associados à representação do sistema de numeração da máquina.
A maioria das máquinas trabalha com o sistema binário. Dessa forma, vamos discutir sobre a a representação de números inteiros e a representação de números reais.

*Representação de números inteiros*:

Um número inteiro na base binária é representado pelos coeficientes de sua expansão binária. Por exemplo,

$13 = 1\times 2^3 + 1\times 2^2 + 0\times 2^1 + 1\times 2^0,$ ou seja, $13 = (1101)_2$.

*Representação de números reais:*

Na base decimal, representamos os números reais,
$$N = a_n10^n+\ldots+a_110^1+a_010^0+a_{-1}10^{-1}+\ldots+a_{-m}10^{-m},$$
com $0\le a_i\le 9$.

Da mesma forma, para qualquer base $b\ge 2$, temos
$$N=\alpha_nb^n+\ldots+\alpha_1b^1+\alpha_0b^0+\alpha_{-1}b^{-1}+\ldots+\alpha_{-m}b^{-m},$$
com $0\le \alpha_i < b$.

**Exemplo 3.**   $9.625=(1001.101)_2$ pois $9.625=1\times 2^3+0\times 2^2 + 0\times 2^1 + 1\times 2^0+1\times 2^{-1} + 0\times 2^{-2}+1\times 2^{-3}$.

*Aritmética de ponto flutuante:*

A Aritmética de ponto flutuante é utilizada em cálculos científicos, padronizada para a representação de números reais. A representação é definida pela base, pela quantidade de algarismos significativos e pelo expoente. A representação de um número $x$ em ponto flutuante, na base $b$, com $n$ algarismos significativos e expoente $e$ é dada por:

$$fl(x) = \pm(.a_1a_2\ldots a_n)\times b^e,$$
sendo $a_1a_2\ldots a_n$ a mantissa do número, com $a_1\neq 0$.

**Exemplo 4.**

(a) Na base decimal os números $0.02531$ e $25.31$, possuem a mesma mantissa e são representados, respectivamente, $0.2531\times 10^{-1}$ e $0.2531\times 10^2$.

(b) Na base binária, dois números consecutivos $0.11100100\times 2^2$ e $0.11100101\times 2^2$, com $8$ algarismos significativos, são respectivamente:

$ 3.5625= \left(\dfrac{1}{2}+\dfrac{1}{2^2}+\dfrac{1}{2^3}+\dfrac{1}{2^6}\right)\times 2^2$ e $3.578125 =  \left(\dfrac{1}{2}+\dfrac{1}{2^2}+\dfrac{1}{2^3}+\dfrac{1}{2^6}+\dfrac{1}{2^8}\right)\times 2^2$

Note que os números entre $3.5625$ e $3.578125$, bem como muitos outros números reais não pode ser representado exatamente.

O mesmo ocorre com qualquer máquina. A representação de um número $x$ em Aritmética de ponto flutuante em um computador com 64 *bits* é dada por:

$$fl(x) = (-1)^{s}+2^{c-1023}(1+m),$$
sendo $s$ o sinal (1 *bit*), $c$ o expoente (11 *bits*) e $m$ a mantissa (52 *bits*).

*Observações:*

* o expoente $c$ com 11 *bits* nos fornece uma variação de $0$ a $2^{11}-1=2047$, que é normalizada de $-1023$ a $1024$.

* Os $52$ *bits* na mantissa equivalem a $16$ algarismos decimais de precisão.

* Em qualquer representação em ponto flutuante temos o menor e o maior número que pode ser representado, indicando *underflow* e o *overflow*.



**Exemplo 5.** Na representação de 64 *bits* descrita acima, o número

Primeiro vamos analisar a variação do expoente. Com 4 *bits* temos $2^4=16$ possibilidades, uma variação de 0 a 15, que após a normalização fica de $-7$ a $8$. Assim, a representação é dada por,

A quantidade de algarismos significativos permanece fixa ao longo de todos os cálculos. Assim temos duas formas de aproximar a mantissa de um número.

* **Corte**: $.a_1a_2\ldots a_ta_{t+1}\ldots \approx .a_1a_2\ldots a_t$, sendo $t$ a quantidade de algarsimos significativos.

* **Arredondamento**: $.a_1a_2\ldots a_ta_{t+1}\ldots \approx .a_1a_2\ldots \tilde{a_t}$, sendo $t$ a quantidade de algarsimos significativos e $\tilde{a_t}=a_t $ ou $\tilde{a_t}=a_t + 1$, dependendo do resultado de $a_{t+1}+5$.

Um sistema de ponto flutuante pode ser expresso pela função $F(b,n,e_1,e_2)$, onde $b$ é a base, $n$ é a quantidade de algarismos significativos da mantissa e $e_1$ e $e_2$ o menor e o maior expoente, respectivamente.

**Exemplo 7.** Representar os números $1.234$ e $1.235$ usando três algarismos significativos.

Usando o corte temos: $1.23$ em ambos os casos.

Usando o arrendondamento temos:

$1.234$ é representado por $1.23$, pois somando $5$ ao próximo algarismo (o quarto) teremos o número $1.239$ não afetanto o terceiro algarismo.

Já $1.235$ é representado por $1.24$, pois ao somarmos $5$ ao quarto algarismo obtemos $1.240$.

**Medidas de erros:**

Se $v$ é uma aproximação do valor exato de $u$, definimos o **erro absoluto** como:

$$e = |u-v|$$

Já o **erro relativo** é dado por
$$e = \dfrac{|u-v|}{|u|}$$

Se $\bf{v},{\bf u}\in\mathbb{R}^n$ então utilizaremos normas para avaliar o erro. As mais utilizadas são:

*Norma 2*: $||{\bf v} - {\bf u}||_2=\sqrt{\left(\sum_{i=1}^n(v_i - u_i)^2\right)}$

*Norma* $\infty$: $\displaystyle{||{\bf v} - {\bf u}||_{\infty}=\max_{1\le i\le n}|v_i - u_i|}$

Como nem sempre conhecemos os valores exatos, ficamos impossibilitados de calcular o erro. Por isso, é importante conhecermos estimativas de erro.

**Lema:** O erro na representação de um número $a\neq 0$ em ponto flutuante com $t$ algarismos significativos é limitado por

$$|a-fl(a)|\le 5|a|10^{-t}p,$$

com $p=1$ (arredondamento) e $p=2$ (corte).

*demonstração:*

Supondo $a=10^q(.d_1d_2\ldots d_td_{t+1}\ldots)$, usando o corte temos $fl(a)=10^q(.d_1\ldots d_t)$. Assim, $|a-fl(a)|=10^{q-t}(.d_{t+1}\ldots)$. Multiplicando o lado direito por $|a|/|a|$ obtemos, $|a-fl(a)|=10^{q-t}(.d_{t+1}\ldots)\dfrac{|a|}{|a|}=10^{-t}\dfrac{(.d_{t+1}\ldots)}{(.d_1\ldots d_t)}|a|$. Como $1\le d_1\le 9$ e $(.d_{t+1}\ldots)\le 1$, temos que quanto menor o denominador, $\dfrac{.d_{t+1}\ldots}{.d_1\ldots d_t}\le \dfrac{1}{0.1}$ maior a estimativa. Portanto,

$$|a-fl(a)|=10^{1-t}|a|.$$
